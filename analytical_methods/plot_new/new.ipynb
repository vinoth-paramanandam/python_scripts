```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df5LnB = pd.read_csv('M5_LnB.csv')
df5LnBMach = pd.read_csv('M5_LnBMach.csv')
df5cfd = pd.read_csv('m5.csv')
```

```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
# fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, gridspec_kw={'width_ratios':[3,3]}, constrained_layout=True)
fig, ax1 = plt.subplots(1, 1)
_ = fig.suptitle('Analytical Mach stem height vs Nozzle Pressure Ratio')

_ = ax1.plot(df5LnB['npr'], df5LnB['Hm'], linestyle='-', color='red')
_ = ax1.plot(df5LnBMach['npr'], df5LnBMach['Hm'], linestyle='-.', color='red')
_ = ax1.plot(df5cfd['npr'], df5cfd['Hm'], marker='*', color='red')
```

```python
df5LnB = pd.read_csv('M5_LnB.csv')
df5LnBMach = pd.read_csv('M5_LnBMach.csv')
df5cfd = pd.read_csv('m5.csv')
df5Mou = pd.read_csv('M5_Mou.csv')
```

```python
fig, ax = plt.subplots(figsize=(6,5))

_ = ax.plot(df5LnBMach['npr'], df5LnBMach['Hm'], linestyle='-', color='red', linewidth=0.75)
_ = ax.plot(df5Mou['npr'], df5Mou['Hm'], linestyle='-.', color='red', linewidth=0.75)
_ = ax.scatter(df5cfd['npr'], df5cfd['Hm'], marker='o', color='red', alpha=0.5, edgecolor='black')

ax.set_xlabel('Nozzle Pressure Ratio')
ax.set_ylabel(r'Machstem Height ($\frac{H_m}{H}$)')
ax.legend(('Li and Bendor Method', 'Mouton and Hornung Method', 'CFD Result'), fontsize='x-small')


ax.set_xlim(35, 75)
ax.set_ylim(0, 0.6)
ax.set_aspect(50)
ax.grid('on', linestyle='-.')

fig.tight_layout()

fig.show()
fig.savefig('example.pdf')
```
