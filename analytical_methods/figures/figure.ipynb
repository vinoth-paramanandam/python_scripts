```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
regime = pd.read_csv('regimes.csv')
```

```python
regime.columns
```

```python
%config InlineBackend.figure_format='retina'
%matplotlib inline
```

```python
plt.rc('font', family='serif', serif='Times')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
plt.rc('axes', labelsize=12)
```

```python
fig, ax = plt.subplots(figsize=(6,5))

_ = ax.plot(regime['M'], regime['second'], linestyle='-', color='red', linewidth=0.75)
_ = ax.plot(regime['M'], regime['sonic'], linestyle='-', color='green', linewidth=0.75)
_ = ax.plot(regime['M'], regime['detach'], linestyle='-', color='blue', linewidth=0.75)
_ = ax.plot(regime['M'], regime['von'], linestyle='-', color='black', linewidth=0.75)
_ = ax.plot(regime['M'], regime['thrid'], linestyle='-', color='magenta', linewidth=0.75)

ax.set_xlabel('Mach Number')
ax.set_ylabel('Nozzle Pressure Ratio')
ax.legend(('Second Critical', 'Sonic', 'Detachment', 'vonNeumann', 'Third Critical'), fontsize='x-small')


ax.set_xlim(1, 5)
ax.set_ylim(1, 300)
ax.set_aspect(0.012)
ax.grid('on', linestyle='-.')

fig.tight_layout()

fig.show()
fig.savefig('example.pdf')
```
